Computing the probability at least one neuronal address is within a given Hamming distance of an address. Assumes neurons are randomly distributed throughout the space and takes an expectation.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm
from scipy.sparse import csc_matrix, coo_matrix, csr_matrix
import pandas as pd
import time
```

```python
n=1000
nprime = 1000000
ds = np.arange(0,n//2)
ys = 1-np.exp( -nprime*norm.cdf(ds, loc=n/2, scale=np.sqrt(n/4)) ) 
plt.plot( ds, ys )
plt.title('Probability of at least one neuronal address \n being within $d$ Hamming distance. $n=1,000$, $r=1,000,000$.')
plt.ylabel('Probability')
plt.xlabel('Hamming distance $d$')
#plt.axhline(0.5, color='red')
plt.axvline(424, color='red')

plt.gcf().savefig('figures/SDMProbAddressWithinDistanceD.png', dpi=250)
plt.show()
```
